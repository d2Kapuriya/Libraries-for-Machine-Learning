#Numpy
NumPy’s main object is the homogeneous multidimensional array.
It is a table of elements (usually numbers), all of the same type, indexed by a tuple of positive integers.

##•	Example of array creation:


```python
import numpy as np
arr = np.array( [[ 1, 2, 3],
                [ 4, 2, 5]] )
print(arr)
```

```python
#creating arrays
np.zeros(10, dtype='int')
```

```python
#creating a 3 row x 5 column matrix
np.ones((3,5), dtype=float)
```

```python
#creating a matrix with a predefined value
np.full((3,5),1.23)
```

```python
#create an array with a set sequence
np.arange(0, 20, 2)
```

```python
#create an array of even space between the given range of values
np.linspace(0, 1, 5)
```

```python
#create a 3x3 array with mean 0 and standard deviation 1 in a given dimension
np.random.normal(0, 1, (3,3))
```

##Array Slicing

**Syntax:** a[start:stop:step]

```python
x = np.arange(10)
x
```

```python
#from start to 4th position
x[:5]
```

```python
#from 4th to 6th position
x[4:7]
```

```python
#return elements at even place
x[ : : 2]
```

##List Comprehension

```python
L = list(range(10))
[str(c) for c in L]
```

##Shape Manipulation
**Syntax:** numpy.reshape(a, newshape)

```python
x = np.array([[2,3,4], [5,6,7]])
np.reshape(x, (3, 2))
```

#Pandas

##Creating DataFrame:


1.   Creating Pandas DataFrame from lists of lists.



```python
import pandas as pd 
# initialize list of lists 
data = [['tom', 10], ['nick', 15], ['juli', 14]] 
# Create the pandas DataFrame 
df = pd.DataFrame(data, columns = ['Name', 'Age']) 
df
```


2.   Creating DataFrame from dict of narray/lists.


```python
import pandas as pd 
# intialise data of lists. 
data = {'Name':['Tom', 'nick', 'krish', 'jack'], 'Age':[20, 21, 19, 18]} 
# Create DataFrame 
df = pd.DataFrame(data) 
df
```



3.   Creates a indexes DataFrame using arrays.



```python
import pandas as pd   
# initialise data of lists. 
data = {'Name':['Tom', 'Jack', 'nick', 'juli'], 'marks':[99, 98, 95, 90]}  
# Creates pandas DataFrame. 
df = pd.DataFrame(data, index =['rank1', 'rank2', 'rank3', 'rank4']) 
df
```
